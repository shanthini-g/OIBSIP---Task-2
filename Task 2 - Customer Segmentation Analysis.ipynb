```python
# DATA EXPLORATION
```

```python
plt.figure(figsize =(10,8))
sns.boxplot(data=df, y='Income', palette='viridis')
plt.title('Box Plot for Income')
plt.ylabel('Income')
plt.show()
```

```python
plt.figure(figsize =(10,8))
sns.boxplot(data=df, y='MntTotal')
plt.title('Box Plot of Total Amount Spent')
plt.ylabel('Total Amount Spent')
plt.show()
```

```python
plt.figure(figsize =(10,8))
sns.histplot(data=df, x='Income', bins = 30, kde=True)
plt.title('Histogram of Income')
plt.xlabel('Income')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize =(10,8))
sns.histplot(data=df, x='Age', bins = 30, kde=True)
plt.title('Histogram for Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# CALCULATETHE CORRELATION MATRIX
correlation_matrix = df.corr()
```

```python
# DISPLAY THE CORRELATION MATRIX
print(correlation_matrix)
```

```python
plt.figure(figsize =(20,12))
sns.heatmap(correlation_matrix, annot = True,fmt = ".2f", cmap = 'coolwarm', linewidths = 0.5)
plt.title('Correlation of Heatmap')
plt.show()
```
